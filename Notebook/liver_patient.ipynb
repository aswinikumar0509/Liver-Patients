```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test) 
```

```python
from imblearn. over_sampling import SMOTE
```

```python
# Resample the training set using SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
y_train_resampled.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import SelectFromModel
```

```python
## Selecting the most important feature 
selector = SelectFromModel(estimator = RandomForestClassifier(random_state=42))
X_train_selected = selector.fit_transform(X_train_resampled,y_train_resampled)
X_test_selected = selector.transform(X_test)
```

```python
xgb = XGBClassifier(random_state=42)
xgb.fit(X_train_selected,y_train_resampled)
y_pred = xgb.predict(X_test_selected)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

param_grid = {
    'learning_rate':[0.01,0.1,0.3],
    'max_depth':[3,5,7],
    'n_estimators':[50,100,200],
    'reg_alpha':[0,0.1,0.5],
    'reg_lambda':[0,1,2]
}

## Creating the Xgboost Classifier

xgb = XGBClassifier(random_state=42)

## Performing the grid search cv for finding the best parameter

grid_search = GridSearchCV(xgb,param_grid = param_grid , cv = 5 , n_jobs = 1 , scoring = 'roc_auc')
grid_search.fit(X_train_selected,y_train_resampled)

print("Best hyperparameter : " , grid_search.best_params_)
print("Best ROC and AUC score : " , grid_search.best_score_)
```

```python
#Evaluate the model with the best hyperparameters on the test data
xgb_best = grid_search.best_estimator_
xgb_best.fit(X_train_selected, y_train_resampled)
y_pred = xgb_best.predict(X_test_selected)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
from sklearn.model_selection import cross_val_score

# Define the XGBoost classifier with the best hyperparameters
xgb = XGBClassifier(learning_rate=0.3, max_depth=7, n_estimators=200, reg_alpha=0.5, reg_lambda=2, random_state=42)

# Evaluate the model performance using 5-fold cross-validation
cv_scores = cross_val_score(xgb, X_train_selected, y_train_resampled, cv=5, scoring='roc_auc')
acc_scores = cross_val_score(xgb, X_train_selected, y_train_resampled, cv=5, scoring='accuracy')

# Print the cross-validation scores and their mean
print('Cross-Validation ROC-AUC Scores: ', cv_scores)
print('Mean Cross-Validation ROC-AUC Score: ', cv_scores.mean())
```
